```python
# Import Necessary libraries

import random
import numpy as np
from typing import List, Tuple, Dict
import time
import matplotlib.pyplot as plt
import os
from google.colab import files

class TaskSchedulerACO:

    # Constructor definition
    def __init__(self, tasks: List[int], vms: List[int], time_matrix: np.ndarray,
                 num_ants: int = 20, iterations: int = 100, alpha: float = 1.0,
                 beta: float = 2.0, rho: float = 0.1, q: float = 10.0):
        self.tasks = tasks # tasks
        self.vms = vms # virtual machines
        self.time_matrix = time_matrix # execution time matrix
        self.num_tasks = len(tasks) # number of tasks
        self.num_vms = len(vms) # number of vm
        self.num_ants = num_ants # number of ants
        self.iterations = iterations # number of iterations
        self.alpha = alpha # weight of pheromone influence
        self.beta = beta # weight of heuristic influence
        self.rho = rho # evaporation rate
        self.q = q # pheromone deposit factor

        # Initialize pheromone matrix with 1s (tasks x vms)
        self.pheromones = np.ones((self.num_tasks, self.num_vms))

        # Heuristic information (inverse of execution time)
        self.heuristic = 1 / (time_matrix + 1e-10)

        # Store makespan values for plotting convergence (best makespan in each iteration)
        self.convergence_history = []

# Construction of solution by assigning each task to a vm
    def construct_solution(self) -> List[int]:
        solution = []

        # Loop through tasks and initialize probability list for each task
        for task_idx in range(self.num_tasks):
            probabilities = []
            total = 0.0

           # Compute probability for each vm based on pheromone and heuristic values
            for vm_idx in range(self.num_vms):
                pheromone = self.pheromones[task_idx][vm_idx]
                heuristic = self.heuristic[task_idx][vm_idx]
                prob = (pheromone ** self.alpha) * (heuristic ** self.beta)
                probabilities.append(prob)
                total += prob

            # Roulette wheel selection, choose vm based on computed probability
            rand = random.uniform(0, total)
            cumulative = 0.0

            # Higher pheromone and lower execution time will have a higher chance of selection
            for vm_idx, prob in enumerate(probabilities):
                cumulative += prob
                if cumulative >= rand:
                    solution.append(vm_idx)
                    break
        return solution

    # Calculate makespan
    def calculate_makespan(self, solution: List[int]) -> Tuple[float, List[float]]:

        # Array vm_times to store execution time of each vm
        vm_times = np.zeros(self.num_vms)

        # Assign each task to a vm and update total execution time for that vm
        for task_idx, vm_idx in enumerate(solution):
            vm_times[vm_idx] += self.time_matrix[task_idx][vm_idx]

        # Return max vm time (makespan) and execution times of all vms
        return np.max(vm_times), vm_times.tolist()

    # Updating pheromones
    def update_pheromones(self, solutions: List[List[int]], makespans: List[float]):

        # Evaporate pheromones
        self.pheromones *= (1 - self.rho)

        # Find best solution (lowest makespan)
        best_idx = np.argmin(makespans)
        best_solution = solutions[best_idx]
        best_makespan = makespans[best_idx]

        # Deposit pheromones on the best path
        for task_idx, vm_idx in enumerate(best_solution):
            self.pheromones[task_idx][vm_idx] += self.q / best_makespan

# Optimization process
    def optimize(self) -> Tuple[List[int], float, List[float]]:

      # Initialize best solution variables
        best_solution = None
        best_makespan = float('inf')
        best_vm_times = []
        self.convergence_history = []

# Generate solutions for all ants
        for _ in range(self.iterations):
            solutions = [self.construct_solution() for _ in range(self.num_ants)]
            makespans = []
            vm_times_list = []

# Compute makespan for each solution
            for sol in solutions:
                makespan, vm_times = self.calculate_makespan(sol)
                makespans.append(makespan)
                vm_times_list.append(vm_times)

            # Update best solution if a better one is found
            current_best = min(makespans)
            if current_best < best_makespan:
                best_makespan = current_best
                best_idx = np.argmin(makespans)
                best_solution = solutions[best_idx]
                best_vm_times = vm_times_list[best_idx]

# Update pheromones and store convergence history

            self.update_pheromones(solutions, makespans)
            self.convergence_history.append(best_makespan)

        return best_solution, best_makespan, best_vm_times, self.convergence_history

# Will read task and vms lists from a file
def parse_input_file(filename: str) -> Tuple[List[int], List[int], np.ndarray]:
    with open(filename, 'r') as f:
        # Read tasks
        tasks = list(map(int, f.readline().strip().split()))

        # Read VMs
        vms = list(map(int, f.readline().strip().replace(',', ' ').split()))

        # Read time entries
        time_entries = f.readline().strip().replace('(', '').replace(')', '').split()

        # Create time matrix
        time_matrix = np.zeros((len(tasks), len(vms)))
        for entry in time_entries:
            parts = list(map(int, entry.split(',')))
            task_idx = tasks.index(parts[0])
            vm_idx = vms.index(parts[1])
            time_matrix[task_idx][vm_idx] = parts[2]

        return tasks, vms, time_matrix

# Generate random execution times for tasks on vm's
def generate_random_instance(n: int, m: int) -> Tuple[List[int], List[int], np.ndarray]:
    tasks = list(range(1, n+1))
    vms = list(range(1, m+1))
    time_matrix = np.random.randint(5, 11, size=(n, m))
    return tasks, vms, time_matrix

# Plot makespan over iterations
def plot_convergence(convergence_data: List[float], title: str = "ACO Convergence"):
    plt.figure(figsize=(10, 6))
    plt.plot(convergence_data, marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel("Iteration")
    plt.ylabel("Makespan")
    plt.grid(True)
    plt.savefig(f"{title.replace(' ', '_')}.png")
    plt.show()
    plt.close()


def run_experiments():
    # Generate and test instances for n=10 to 20, m=5
    results = []

    for n in range(10, 21):
        m = 5
        try:
            # Generate input file
            filename = f"experiment_n{n}_m{m}.txt"
            generate_input_file(filename, n, m)

            # Parse generated file
            tasks, vms, time_matrix = parse_input_file(filename)

            start_time = time.time()
            # run ACO
            aco = TaskSchedulerACO(tasks, vms, time_matrix)
            solution, makespan, vm_times, convergence = aco.optimize()
            execution_time = time.time() - start_time

            # Plot convergence
            plot_title = f"ACO Convergence (n={n}, m={m})"
            plot_convergence(convergence, plot_title)

            # Calculate metrics
            avg_vm_utilization = np.mean(vm_times) / makespan * 100
            load_balance = np.std(vm_times) / makespan * 100

            results.append({
                'n': n,
                'm': m,
                'solution': solution,
                'makespan': makespan,
                'vm_times': vm_times,
                'execution_time': execution_time,
                'avg_utilization': avg_vm_utilization,
                'load_balance': load_balance
            })

            # Print results
            print(f"\nResults for n={n}, m={m}:")
            print(f"Optimal Assignment: {solution}")
            print(f"Makespan: {makespan}")
            print(f"VM Times: {vm_times}")
            print(f"Execution Time: {execution_time:.2f}s")
            print(f"Average VM Utilization: {avg_vm_utilization:.2f}%")
            print(f"Load Balance (STD%): {load_balance:.2f}%")

        except Exception as e:
            print(f" Error processing n={n}: {str(e)}")

    # Performance analysis
    print("\n\n Performance Summary:\n")
    print("n\tMakespan\tAvg Util(%)\tBalance(%)\tTime(s)")
    for res in results:
        print(f"{res['n']}\t{res['makespan']}\t\t{res['avg_utilization']:.1f}\t\t"
              f"{res['load_balance']:.1f}\t\t{res['execution_time']:.2f}")


def generate_input_file(filename: str, n_tasks: int, n_vms: int):
    """Generate a benchmark file with random task-VM times"""
    # Validate input
    if n_vms >= n_tasks:
        raise ValueError("Number of VMs must be less than number of tasks (m < n)")

    # Generate tasks and VMs
    tasks = list(range(1, n_tasks+1))
    vms = list(range(1, n_vms+1))

    # Generate random times (5-10 units) for all task-VM combinations
    time_entries = []
    for task in tasks:
        for vm in vms:
            time = random.randint(5, 10)
            time_entries.append(f"({task},{vm},{time})")

    # Write to file
    with open(filename, 'w') as f:
        f.write(' '.join(map(str, tasks)) + '\n')
        f.write(' '.join(map(str, vms)) + '\n')
        f.write(' '.join(time_entries))

    files.download(filename)

def main():
    # Generate experiment files and run tests
    print("\n Running experiments...")
    run_experiments()




if __name__ == "__main__":
    main()
```
